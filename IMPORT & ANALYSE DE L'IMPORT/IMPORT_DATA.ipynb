```python
import pymongo
import requests
```

```python
# Connexion à la base de données MongoDB
client = pymongo.MongoClient("mongodb://root:GFUGFISIUFBEIUGFIUSUF@localhost:27017/")
```

```python
db = client["tgvmax"]
collection = db["data"]

# Paramètres de pagination
page = 1
limit = 100

while True:
    # Récupération des données de l'API SNCF
    response = requests.get(f"https://ressources.data.sncf.com/api/explore/v2.1/catalog/datasets/tgvmax/records?page={page}&limit={limit}")
    data = response.json()
    results = data.get("results", [])

    if not results:
        break

    # Insérer les données dans la collection
    collection.insert_many(results)

    print(f"Page {page} : {len(results)} enregistrements insérés dans MongoDB.")

    page += 1
```
